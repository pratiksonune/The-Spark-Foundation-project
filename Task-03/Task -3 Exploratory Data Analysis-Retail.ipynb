```python
plt.figure(figsize=(14,14))
sns.countplot(x=SampleSuperstore1['State'])
plt.xticks(rotation=80)
plt.show()
```

```python
from ggplot import *
profit_plot =(ggplot(sample, aes(x='sub-category' ,y='profit',file='sub-category')) +geom_col()+coord_flip()+
              +scale_fill_brewer(type='div',palette="Spectral")+theme_classic()+ggtitle('Pie chart'))

display(profit_plot)
```

```python
sns.set(style ="whitegrid")
plt.figure(2, figsize= (20,15))
sns.barplot(x='Sub-Category',y='Profit',data =SampleSuperstore,palette ='Spectral')
plt.suptitle('Pie Consumption Patterns in the united stste',fontsize = 16)
plt.show()
```

```python
figsize=(15,10)
sns.pairplot(SampleSuperstore1,hue='Sub-Category')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot('Discount','Profit',data = SampleSuperstore1,color='g',label='Discount')
plt.legend()
plt.show()
```
